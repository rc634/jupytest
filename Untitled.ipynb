```python
import torch
from pyannote.database.util import load_rttm

base = "cobal"#"zidwg"#"cobal"

file = "data/"+base+".wav"
truth = "data/"+base+".rttm"

```

```python
groundtruth = load_rttm(truth)[base]
print(groundtruth)
```

```python

groundtruth
```

```python
#speaker activity detection
sad = torch.hub.load("pyannote/pyannote-audio","sad")
speech_scores = sad({"audio":file})
speech_scores
```

```python
#speaker change detection
scd = torch.hub.load("pyannote/pyannote-audio","scd")
change_scores = scd({"audio":file})
change_scores
```

```python
#overlapped speech detection
ovl= torch.hub.load("pyannote/pyannote-audio","ovl")
ovl_scores = ovl({"audio":file})
ovl_scores
```

```python
diarization = torch.hub.load("pyannote/pyannote-audio","dia")
output = diarization({"audio":file})
output
```

```python
with open(truth,"w") as truthfile:
    output.write_rttm(truthfile)
```

```python
# instantiate pretrained speaker diarization pipeline
import pyannote.audio
from pyannote.audio import pipeline
ppipeline = pipeline.from_pretrained("pyannote/speaker-diarization")

# apply pretrained pipeline
diarization = ppipeline(file)

# print the result
for turn, _, speaker in diarization.itertracks(yield_label=True):
    print(f"start={turn.start:.1f}s stop={turn.end:.1f}s speaker_{speaker}")
# start=0.2s stop=1.5s speaker_A
# start=1.8s stop=3.9s speaker_B
# start=4.2s stop=5.7s speaker_A
# ...
```
